```python
import pandas as pd
from plotly.graph_objs import *
from plotly.offline import init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
```

```python
dataset = pd.read_csv('finalDataset.csv')
```

```python
dataset
```

```python
yearlist = [1990,1995,2000,2005,2010,2014]
genrelist = ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 'History', 'Horror',
 'Musical', 'Mystery' ,'Romance', 'Sci-Fi', 'Sport', 'Thriller', 'War', 'Western']
```

```python
def ratingbygenre(genre,year):
    return round(dataset[dataset.GENRE.str.contains(genre, na = False) & (dataset.YEAR == year)] ['RATING'].mean(),2)
```

```python
list1990 = [ratingbygenre(genre, year[0]) for genre in genrelist]
#list1990
```

```python
list1995 = [ratingbygenre(genre, year[1]) for genre in genrelist]
#list1995
```

```python
list2000 = [ratingbygenre(genre, year[2]) for genre in genrelist]
list2005 = [ratingbygenre(genre, year[3]) for genre in genrelist]
list2010 = [ratingbygenre(genre, year[4]) for genre in genrelist]
list2014 = [ratingbygenre(genre, year[5]) for genre in genrelist]
```

```python
ratinggenrelist = [meanrating_action,meanrating_adventure,meanrating_animation,meanrating_biography,meanrating_comedy,
                   meanrating_crime,meanrating_documentary,meanrating_drama,meanrating_family,
                   meanrating_fantasy,meanrating_history,meanrating_horror,meanrating_musical,meanrating_mystery,
                    meanrating_romance,meanrating_scifi,meanrating_sport,meanrating_thriller,meanrating_war,meanrating_western]

ratinggenrelist
```
